### 아래의 성적 파일을 사용하여 각 문제를 푸시오.

```python
import pandas as pd

df_grade = pd.read_excel("grade.xlsx")
df_grade
```

#### 1. 각 학생에 대하여 총점을 추가하시오. 이 때, 컬럼명은 '총점'으로 한다.

```python
subjects = ["국어","영어","수학","과학"]
df_grade["총점"] = df_grade[subjects].sum(axis=1)
df_grade
```

#### 2. 각 학생에 대하여 평균을 추가하시오. 이 때, 컬럼명은 '평균'으로 하며 소수점 첫 째자리까지만 표시한다.

```python
df_grade["평균"] = round(df_grade["총점"] / len(subjects), 1)
df_grade
```

####  3. 전체 학생에 대하여 평균을 내림차순으로 정렬하시오.

```python
df_grade.sort_values("평균", ascending=False)
```

#### 4. # 3. 평균에 대하여 다음의 등급을 추가하시오. 이 때, 컬럼명은 '등급'으로 한다.
1. 100~91: A
2. 90~81: B
3. 80~71: C
4. 70~61: D
5. 60~51: E
6. 50~0: F

```python
def mean_to_grade(mean):
    if mean > 90:
        return "A"
    elif mean > 80:
        return "B"
    elif mean > 70:
        return "C"
    elif mean > 60:
        return "D"
    elif mean > 50:
        return "E"
    else:
        return "F"
    
df_grade["등급"] = df_grade["평균"].apply(mean_to_grade)
df_grade
```

#### 5. 성적 데이터프레임에서 결측치의 총 개수를 출력하시오.

```python
print(df_grade.isna().sum().sum())
```
